```python
import numpy as np
import pandas as pd
```

```python
#Set values for the run
prime = 2 
beta = -.99
set_charges = [1,-1,1,-1]
```

```python
#Creating Dataframs
df = pd.read_csv("size4p2.csv", delimiter ="|")
df_out = pd.DataFrame(columns = ['tree_id', 'Terms', 'phys_prob'])
df_out['tree_id'] = df['tree_id']
df_out['Terms'] = 0.0
df_out['phys_prob'] = 0
df_out
```

```python
def falling_factorial(x,n):
    prod = 1.0
    for i in range (0, n):
        prod *= x-i
    return prod
```

```python
def term(branches, degrees, p, beta, intrc_enrgy_array):
    #branch_arr = branches.apply(lambda b: list(map(int, b[1:-1].split(","))))
    branch_arr = list(map(int, branches[1:-1].split(",")))
    #deg_arr = degrees.apply(lambda d: list(map(int, d[1:-1].split(","))))
    deg_arr = list(map(int, degrees[1:-1].split(",")))
    out = 1.0
    for i in range(0, len(branch_arr)):
        out *= falling_factorial(p, deg_arr[i])/(p**(bin(branch_arr[i]).count("1") + (beta * intrc_enrgy_array[branch_arr[i]])) - p)
    return out
```

```python
def interaction_energy(charges, size_of_charge): 
    out = np.empty(pow(2, size_of_charge), dtype= float)
    for J in range (0,  pow(2, size_of_charge)):
        sum1 = 0.0
        sum2 = 0.0
        for i in range (1, size_of_charge + 1):
            sum1 += ((bool(J & (1 << (i - 1)))) * charges[i - 1])
            sum2 += ((bool(J & (1 << (i - 1)))) * charges[i - 1] * charges[i - 1])
        out[J] = (((sum1 * sum1) - sum2) / 2.0)
    return out
```

```python
interaction_energy([1,-1,1,-1], 4)
```

```python
for i in range(0, 26):
    df_out.loc[i,"Terms"] = (term(df.loc[i,'branches'], df.loc[i,'degrees'], prime, beta, interaction_energy(set_charges,len(set_charges))))
```

```python
sum = 0.0
for i in range(0, 26):
    sum += df_out.loc[i,"Terms"]
print(sum)
```

```python
df_out['phys_prob'] = df_out['Terms']/sum
display(df_out)
prob_sum = 0.0
for i in range(0, 26):
    prob_sum += df_out.loc[i,"phys_prob"]
print("The sum of the probabilities is")
print(prob_sum)
```
